# Operators in Python

- Arithmetic Operators
- Relational Operators
- Logical Operators
- Bitwise Operators

- Membership Operators

```python
import warnings

warnings.filterwarnings('ignore')

# Your code goes here
```

```python
# Arithmetric Operators
print(5+6)

print(5-6)

print(5*6)

print(5/2)

# return absolute value -> not in floats
print(5//2)
# return remainder
print(5%2)
# square
print(5**2)
```

```python
# Relational Operators
print(4>5)

print(4<5)

print(4>=4)

print(4<=4)

print(4==4)

print(4!=4)
```

```python
# Logical Operators
# both condition should be 1 in order to print 1
print(1 and 0)

# any of them
print(1 or 0)
# 0 it mean false -> convert 0 to 1 and 1 to 0
print(not 1)
```

```python
# Bitwise Operators

# bitwise and
print(2 & 3)

# bitwise or
print(2 | 3)

# bitwise xor
print(2 ^ 3)

# 110 is bit pattern of 3-> convert 0 into 1 and 1 into 0->001 which is 1
print(~3)
# shift binaries of 4 to left side with 2
print(4 >> 2)
# shift binaries of 5 to right side with 2
print(5 << 2)
```

```python
# Assignment Operators

# =
# a = 2

a = 2

# a = a % 2
# a %= 2

a=a+2
a+=2

# a++ ++a

print(a)
```

```python
# Membership Operators

# in/not in

print('D' not in 'Delhi')
print(1 in [1,2,3,4,5,6])
print(1 not in [1,2,3])
print(1 in [2,3,4,5,6])
```

# If-else in Python

```python
# login program and indentation -> if-else=> branching
# email -> Hannan@@gmail.com
# password -> 1234

email=input('enter email')
password=input('enter password')

if email=='Hannan@gmail.com' and password=='1234':
  print('Welcome')

elif email=='Hannan@gmail.com' and password!='1234':
  # tell the user
  print('Wrong password')
  password=input('enter password again')
  if password=='1234':
    print('welcome')

  else:
    print('Try again')

else:
  print('Not correct')
```

# Loops in Python

```python
# While loop example -> program to print the table
# Program -> Sum of all digits of a given number
# Program -> keep accepting numbers from users till he/she enters a 0 and then find the avg
```

```python
number = int(input('enter the number'))

i = 1

while i<11:
  print(number,'*',i,'=',number * i)
  i += 1
```

```python
x=1

while x<3:
  print(x)
  x+=1

else:
  print('limit crossed')
```

```python
# while loop with else

x = 1

while x < 3:
  print(x)
  x += 1

else:
  print('limit crossed')
```

```python
# Guessing game

# generate a random integer between 1 and 100
import random
jackpot = random.randint(1,100)

guess = int(input('guess karo'))
counter = 1
while guess != jackpot:
  if guess < jackpot:
    print('galat!guess higher')
  else:
    print('galat!guess lower')

  guess = int(input('guess karo'))
  counter += 1

else:
  print('correct guess')
  print('attempts',counter)


```

### Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

```python
curr_pop = 10000

for i in range(10,0,-1):
  print(i,curr_pop)
  curr_pop = curr_pop/1.1
```
